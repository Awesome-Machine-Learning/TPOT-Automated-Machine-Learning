### Load the Iris dataset

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

### Make a dataframe

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df
```

### add target column into dataframe

```python
df['target']=iris.target
df
```

---
### Select features n target from the dataset

```python
# select features
X=df[iris.feature_names]

# convert itno array
X=np.asarray(X)
X[0:10]
```

```python
# select arget
y=df['target']
y
```

---
## Split the data into train n test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=2)
```

### Shape of the train n test

```python
X_train.shape,  X_test.shape, y_train.shape,    y_test.shape
```

---
# TPOTClassifier

```python
from tpot import TPOTClassifier
```

```python
tpot=TPOTClassifier(verbosity=2,max_time_mins=10)  # run for 5 mins
# fit the train data
tpot.fit(X_train,y_train)
```

```python
print("Accuracy : ",tpot.score(X_test,y_test))
```

```python
tpot.export("TPOT_iris_pipeline.py")
```
